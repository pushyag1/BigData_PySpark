# Automating for each model year and version

```
for version in versions:
  for year in model_years:
    
    #Part 1: Creating custom dataframe
    df_temp= df_CC_mapping.loc[(df_CC_mapping['VERSION']==str(version)) & (df_CC_mapping['MODEL YEAR']== year)]  #creating custom dataframe based on given version and model year from CC_Mapping dataframe
    

    #Part 2: Mapping Diag codes from Diagnosis input file(diag_list) with diag codes from CC Mapping file
    CC_list=[]
    for i in (diag_list):
      x= df_temp.loc[i==df_temp['DIAGNOSIS CODE']]['CC'].tolist() # Searching where value from list equals row value in dataframe and copying corresponding CC value to list
      CC_list.append(x)
    # print(CC_list)


    #Part 3: Creating new dataframe for each instance 
    df_main= pd.DataFrame()
    df_main['PERSON ID']= list(df_diagnosis['PERSON ID'])
    df_main['DIAGNOSIS CODE']= list(df_diagnosis['DIAGNOSIS CODE']) 
    df_main['HCC']= CC_list  # CC values obtained after mapping from Part 2.
    df_main['MODEL YEAR']= str(year)
    df_main['Version']= str(version)

    #Part 4: Finding all possible HCC values from CC Mapping file
    h= df_temp['CC'].unique().tolist()
    unique_HCC= sorted(h)
    # print(unique_HCC)

    # Part 5: Creating new columns with column names as HCC values obtained from Part 4
    columns_df=[]
    for value in (unique_HCC):
      columns_df.append('HCC'+str(value)) # Example: HCC1,HCC2,HCC3 etc
    #print(columns_df)
    for col in range(len(columns_df)):
      df_main[columns_df[col]]=0  #adding zero's to every column 

    #Part 6: Turning 0's to 1's at corresponding HCC columns

    for v in range(len(df_main)):                          # here v is used as index value
      for v2 in range(len(df_main.iloc[v][2])):            # Indicating each row value in HCC column. Each cell in the HCC column contains a list element. Ex: [18],[18,108],[18,106,108].
        df_main.at[v,'HCC'+ str(df_main.iloc[v][2][v2])]=1  # the str part outputs the value of HCC present in each list. v2 is basically iterating through the lenght of list in each row.


    #Part 5: Turning list objects in the HCC column to string objects
    df_main['HCC'] = [','.join(map(str, l)) for l in df_main['HCC']]    # Ex: turning [18,108] to 18,108

    #Part 6: Generating individual excel sheets for each model year and version
    df_main.to_excel(str(version)+'_'+str(year)+'.xlsx') #Outputting excel files for each scenario
```
